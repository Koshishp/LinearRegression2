```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import linear_model
```

```python
df=pd.read_csv('/Users/bimal/Downloads/canada_per_capita_income.csv')
df.tail()
```

```python
df.isnull().sum()
```

```python
sns.histplot(x='per capita income (US$)',data=df)
```

```python
sns.relplot(x='year',y='per capita income (US$)',data=df)
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['year']], df['per capita income (US$)'])
```

```python
reg.predict(np.array([2020]).reshape(-1, 1))
```

```python
reg.score(df[['year']], df['per capita income (US$)'])
```
